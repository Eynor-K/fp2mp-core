# fp2mp_core — Пример: пешеходные улицы Петроградского района

Демонстрация работы мультиагентной системы.  
Конфигурация берётся из `.env` (требуется `FP2MP_CHAT_URL` и `FP2MP_API_KEY`).

```python
MODEL = "gpt-4o-mini"          # модель через FP2MP_CHAT_URL
MAX_ITERATIONS = 5

QUESTION = (
    "Где построить новую станцию метро в Екатеринбурге?"
)
```

```python
from fp2mp_core.graph import run

result = run(input=QUESTION, model=MODEL, max_iterations=MAX_ITERATIONS)
```

```python
from IPython.display import Markdown, display

display(Markdown(result["output"]))
```

```python
print(f"Сообщений в логе: {len(result['log'])}")
for msg in result["log"]:
    if msg.type == "human":
        print(f"👤 {msg.content[:80]}")
    elif msg.type == "ai" and getattr(msg, "tool_calls", None):
        tc = msg.tool_calls[0]
        tool_input = str(tc.get("args", {}).get("input", ""))
        print(f"🔧 [{msg.name}] {tc['name']}({tool_input[:60]})")
    elif msg.type == "tool":
        print(f"   ↳ {msg.content[:100]}")
    elif msg.type == "ai":
        name = getattr(msg, "name", "") or ""
        print(f"🤖 [{name}] {msg.content[:100]}")
```
