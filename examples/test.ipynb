# fp2mp_core — Пример: пешеходные улицы Петроградского района

Демонстрация работы мультиагентной системы.  
Конфигурация берётся из `.env` (требуется `FP2MP_CHAT_URL` и `FP2MP_API_KEY`).

```python
MODEL = "gpt-4o-mini"          # модель через FP2MP_CHAT_URL
MAX_ITERATIONS = 3

QUESTION = "Где построить новую станцию метро в Екатеринбурге?"
```

```python
from fp2mp_core.graph import run

result = run(input=QUESTION, model=MODEL, max_iterations=MAX_ITERATIONS)
```

```python
from IPython.display import Markdown, display

print(result["input"])   # исходный вопрос
print(result["output"])  # финальный ответ (str)
print(len(result["log"]))  # количество сообщений
```

```python
total_tokens = 0

for i, msg in enumerate(result["log"], start=1):
    is_ai = type(msg).__name__ == "AIMessage"
    n = (msg.usage_metadata or {}).get("total_tokens", 0) if is_ai else 0
    total_tokens += n
    token_info = f" | {n} tokens" if is_ai else ""
    print(f"\n--- message {i}: {type(msg).__name__}{token_info} ---")
    print(msg.content)

print(f"\n{'='*60}")
print(f"Total AIMessage tokens: {total_tokens}")
```

```python
# === BlocksNetAgent: детальный лог инструментов ===
# Внимание: этот low-level вызов повторно запускает граф и тратит токены.
from fp2mp_core.graph import build_graph
from fp2mp_core.state import create_initial_state
from fp2mp_core.llm import set_active_model

set_active_model(MODEL)
graph = build_graph()
initial = create_initial_state(QUESTION, max_iterations=MAX_ITERATIONS)
full_state = graph.invoke(initial)

bn_entries = [e for e in full_state.get("raw_data", []) if e.get("agent") == "BlocksNetAgent"]

if not bn_entries:
    print("BlocksNetAgent не запускался или не оставил записей в raw_data")
else:
    for i, entry in enumerate(bn_entries, 1):
        sq_id = entry.get("sub_query_id", "?")
        conf = entry.get("confidence", 0)
        print(f"\n{'='*60}")
        print(f"BlocksNetAgent entry {i} | sq={sq_id} | conf={conf:.2f}")
        print(f"{'='*60}")

        traces = entry.get("tool_trace") or []
        for trace in traces:
            directive = trace.get("directive", {})
            print(f"\nВопрос: {directive.get('question', '(нет)')[:200]}")
            steps = trace.get("intermediate_steps") or []
            if not steps:
                print("  [нет шагов инструментов — tool_trace пустой]")
            for j, step in enumerate(steps, 1):
                print(f"\n  Шаг {j}: {step.get('tool', '?')}")
                print(f"  Вход: {str(step.get('tool_input', ''))[:300]}")
                obs = str(step.get('observation', ''))
                print(f"  Результат: {obs[:500]}")
            raw = trace.get("raw_output", "")
            if raw:
                print(f"\n  Финальный вывод агента:\n{raw}")

        content = entry.get("content", "")
        if content:
            print(f"\nИтоговый ответ BlocksNetAgent:\n{content[:1000]}")
```

```python
# === CodeSpatialAgent: детальный лог инструментов ===
spatial_entries = [
    e for e in full_state.get("raw_data", [])
    if e.get("agent") == "CodeSpatialAgent"
]

if not spatial_entries:
    print("CodeSpatialAgent не запускался")
else:
    for i, entry in enumerate(spatial_entries, 1):
        sq_id = entry.get("sub_query_id", "?")
        conf = entry.get("confidence", 0)
        print(f"\n{'='*60}")
        print(f"CodeSpatialAgent entry {i} | sq={sq_id} | conf={conf:.2f}")
        print(f"{'='*60}")

        traces = entry.get("tool_trace") or []
        for trace in traces:
            directive = trace.get("directive", {})
            print(f"\nВопрос: {directive.get('question', '(нет)')[:200]}")
            steps = trace.get("intermediate_steps") or []
            if not steps:
                print("  [нет шагов — tool_trace пустой (баг или агент не выполнял код)]")
            for j, step in enumerate(steps, 1):
                print(f"\n  Шаг {j}: {step.get('tool', '?')}")
                code = str(step.get('tool_input', ''))
                print(f"  Код ({len(code)} симв):\n    {code[:400]}")
                obs = str(step.get('observation', ''))
                print(f"  Вывод:\n    {obs[:600]}")
            raw = trace.get("raw_output", "")
            if raw:
                print(f"\n  Финальный вывод агента:\n{raw}")

        content = entry.get("content", "")
        if content:
            print(f"\nИтоговый ответ CodeSpatialAgent:\n{content[:1000]}")
```
